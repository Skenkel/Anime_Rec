```python
scores_non_zero1 =pd.read_csv('mal_scores_train_nonzero_v2.csv')
scores_non_zero2 =pd.read_csv('mal_scores_test_nonzero_v2.csv')
```

```python
scores_non_zero=pd.concat([scores_non_zero1,scores_non_zero2],ignore_index=True)
```

```python
useravg =scores_non_zero.groupby('userid').mean() # this gives me a dataframe with each user's avg score. 
animeavg =scores_non_zero.groupby('animeid').mean() # this gives me a dataframe with each user's avg score. 
```

```python
useravgdict = {} # This is an empty dictionary, which we will use to store a user's name as the key, and their avg score as the value
animeavgdict = {} # This is an empty dictionary, which we will use to store an anime id as the key, and it's avg score as the value
```

```python
for user in useravg.index.values:
    score=useravg.loc[user,'score']
    useravgdict[user]=score
for anime in animeavg.index.values:
    score=animeavg.loc[anime,'score']
    animeavgdict[anime]=score
```

```python
useravgdict # checking to make sure our dict looks good
```

```python
animeavgdict # checking to make sure our dict looks good
```

```python
np.save('anime_avgs.npy', animeavgdict) 
```

```python
scores_non_zero['useravg']= scores_non_zero.userid.map(useravgdict)
scores_zero['useravg']= scores_zero.userid.map(useravgdict)
scores_non_zero['animeavg']= scores_non_zero.animeid.map(animeavgdict)
scores_zero['animeavg']= scores_zero.animeid.map(animeavgdict)
```

```python
scores_non_zero['score_anime_scaled']=scores_non_zero['score']-scores_non_zero['animeavg'] # making a column of how much more or less than avg a user liked an anime
scores_non_zero['score_usr_scaled']=scores_non_zero['score']-scores_non_zero['useravg'] # making a column of how much more or less than avg a user liked an anime
```

V2: Improvements
Users that have 0 scores in the train set, we need to fix that (their embeddings will be random, and the neural net won't ever learn anything about them).

Even though this is very few (30k) scores, it will add noise into the models. 
Given how close some of the neural nets scores are to each other, that noise should be reduced if at all possible. 

```python
ratings_score_train = pd.read_csv('mal_scores_train_nonzero.csv') 
ratings_score_test = pd.read_csv('mal_scores_test_nonzero.csv')
ratings_no_score_train = pd.read_csv('mal_scores_train_zero.csv') 
```

```python
ratings_score_test.shape
```

```python
ratings_score_test_v2 =ratings_score_test[ratings_score_test['userid'].isin(ratings_score_train.userid.values)]
```

```python
ratings_score_train_v2 =ratings_score_test[~ratings_score_test['userid'].isin(ratings_score_train.userid.values)]
```

```python
ratings_score_train_v2
```

```python
concatlist = [ratings_score_train_v2,ratings_score_train] # taking the scores that would have added noise and adding them into the the training data will improve the models. 
ratings_train_to_csv =pd.concat(concatlist,ignore_index=True)
```
